# Asian option analytical solution
This notebook contains the analytical solution for the Asian option. The solution is based on the following references:
- [Kemna, A. G. Z., & Vorst, A. C. F. (1990)](https://ideas.repec.org/a/eee/jbfina/v14y1990i1p113-129.html)
- [Mkhize, N. (2007)](http://hdl.handle.net/10413/429)

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
r = 0.008
sigma = 0.2
s_t = 100
T = 30
K = 100
T_0 = 0
t = 0
```

kemna vorst p 11

```python
def kemna_vorst(s_t=100, K=100, r=0.008, sigma=0.2, T=30):
    d_star = (1/2)*(r-(sigma**2) / 6)*T
    d = (np.log(s_t/K) + (1/2)*(r - (sigma**2) / 6)*T)/(sigma*np.sqrt(T/3))
    C_val = np.exp(-r*T)*(np.exp(d_star)*s_t*norm.cdf(d) - K*norm.cdf(d-sigma*np.sqrt(T/3)))
    return C_val
```

Mkhize p 70, continuous fixed strike geometric average call option

```python
def geometric_mkhize_general(s_t=100, K=100, r=0.008, sigma=0.2, T=30, t=0):
    s_wave_t = s_t ** ((T-t)/T)
    d1 = (T* np.log(s_wave_t/K) + 0.5 * (r - 0.5 * sigma**2) * (T-t)**2 + (sigma**2 * (T-t)**3)/(3*T))/(sigma * np.sqrt(((T-t)**3)/3))
    d2 = (T * np.log(s_wave_t/K) + 0.5 * (r - 0.5 * sigma**2) * (T-t)**2)/(sigma * np.sqrt(((T-t)**3)/3))
    C_G = np.exp(-r*(T-t)) * (s_t * np.exp(((r-0.5*sigma**2)*((T-t)**2))/(2*T) + (sigma**2*((T-t)**3))/(6*T**2))*norm.cdf(d1) - K*norm.cdf(d2))
    P_G = np.exp(-r*(T-t)) * (K*norm.cdf(-d2) - s_t * np.exp((r-0.5*sigma**2)*((T-t)**2)/(2*T) + (sigma**2*(T-t)**3)/(6*T**2))*norm.cdf(-d1))
    return C_G, P_G
```

Mkhize p 69, continuous fixed strike geometric average call option, at t = T_0, should be the same as the previous one

```python
def geometric_mkhize(s_t=100, K=100, r=0.008, sigma=0.2, T=30):
    d_1 = (np.log(s_t/K) + 0.5*(r - (sigma**2)/6)*(T-T_0))/(sigma*np.sqrt(T/3))
    d_2 = d_1 - sigma*np.sqrt(T/3)
    C_0 = s_t* np.exp(-0.5 * (r + (sigma**2)/6)*T)*norm.cdf(d_1) - K*np.exp(-r*T)*norm.cdf(d_2)
    P_0 = K*np.exp(-r*T)*norm.cdf(-d_2) - s_t*np.exp(-0.5*(r + (sigma**2)/6)*T)*norm.cdf(-d_1)
    return C_0, P_0
```

```python
assert np.allclose(geometric_mkhize(), geometric_mkhize_general(t=0), rtol=1e-5)
```

```python
K_s = np.linspace(80, 120, 100)
C_kemna = []
C_geometric = []
for K in K_s:
    C_kemna.append(kemna_vorst(K=K))
    C_geometric.append(geometric_mkhize(K=K)[0])

plt.plot(K_s, C_kemna, label='kemna vorst', color='red')
plt.plot(K_s, C_geometric, label='geometric', color='blue')
plt.legend()
plt.xlabel('strike price')
plt.ylabel('option price')
plt.show()
```

Kemna Vorst p 12, asian is allways cheaper or equal to european

```python
def european_option(s_t=100, K=100, r=0.008, sigma=0.2, T=30):
    d_1 = (np.log(s_t/K) + (r + 0.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d_2 = d_1 - sigma*np.sqrt(T)
    C_0 = s_t*norm.cdf(d_1) - K*np.exp(-r*T)*norm.cdf(d_2)
    P_0 = K*np.exp(-r*T)*norm.cdf(-d_2) - s_t*norm.cdf(-d_1)
    return C_0, P_0
```
